FAKE OR REAL NEWS USING NLP

```python
import pandas as pd 
import numpy as np 
import spacy 
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
df = pd.read_csv('Fake_Real_Data.csv')
df.head()
```

```python
df.shape
```

```python
df.label.value_counts()
```

```python
df['Label_num'] = df['label'].map({'Fake':0,'Real':1})
df.head()
```

```python
nlp = spacy.load('en_core_web_lg')
doc = nlp("Top Trump Surrogate BRUTALLY Stabs Him In The..")
doc.vector.shape
```

```python
df['vector'] = df['Text'].apply(lambda x: nlp(x).vector)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split 

X_train,X_test,Y_train,Y_test = train_test_split(
    df.vector,
    df.Label_num,
    test_size = 0.2,
    random_state=2022
)
```

```python
X_train
```

```python
X_test
```

```python
import numpy as np 
X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)
```

```python
X_train_2d
```

```python
X_test_2d
```

Scaling is done to remove the negative values so for that you use MINMAXSCALAR

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler

clf = MultinomialNB()
scalar = MinMaxScaler()

scaled_train = scalar.fit_transform(X_train_2d)
scaled_test = scalar.fit_transform(X_test_2d) 

clf.fit(scaled_train,Y_train)
```

```python
y_pred = clf.predict(scaled_test)
print(classification_report(Y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier 
clf =KNeighborsClassifier(n_neighbors=5,metric ='euclidean')
clf.fit(X_train_2d,Y_train)
y_pred = clf.predict(X_test_2d)

print(classification_report(Y_test,y_pred))
```
